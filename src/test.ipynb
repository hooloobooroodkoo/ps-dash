```python
import plotly_express as px
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output

tips = px.data.tips()
col_options = [dict(label=x, value=x) for x in tips.columns]
dimensions = ["x", "y", "color", "facet_col", "facet_row"]

indexes = ["packet loss", "one way delay"]
df = LoadBubbleChartData(isDev, 'ps_packetloss')
col_options1 = [dict(label=x, value=x) for x in df.columns]

app = dash.Dash(
    __name__, external_stylesheets=["https://codepen.io/chriddyp/pen/bWLwgP.css"]
)

app.layout = html.Div(
    [
#         html.H1("Demo: Plotly Express in Dash with Tips Dataset"),
        html.Div(
            [
                html.P([d + ":", dcc.Dropdown(id=d, options=col_options)])
                for d in dimensions
            ],
            style={"width": "25%", "float": "left"},
        ),
        dcc.Graph(id="graph", style={"width": "75%", "display": "inline-block"}),
        
#         html.Div(
#             [
#                 html.P([d + ":", dcc.Dropdown(id=d, options=col_options1)])
#                 for d in indexes
#             ],
#             style={"width": "20%", "float": "left"},
#         ),
#         dcc.Graph(id="bubble-plot", style={"width": "80%", "display": "inline-block"}),
        
        html.Div(
            dcc.Graph(figure=px.scatter(df, x="period", y="host", 
           size=df["mean"].fillna(value=0), color="host", hover_name="host", size_max=45, height=700)
                  .update(layout={
                      'title':'Avg Packet Loss from 01-12-2019 to 22-01-2020',
                      'xaxis':{'title':'Period'},
                      'yaxis':{'title':'Hosts'},
                      'paper_bgcolor':'rgba(0,0,0,0)',
                      'plot_bgcolor':'rgba(0,0,0,0)'
                  }),
                 ),)
    ]
)


@app.callback(Output("graph", "figure"), [Input(d, "value") for d in dimensions])
def make_figure(x, y, color, facet_col, facet_row):
    return px.scatter(
        tips,
        x=x,
        y=y,
        color=color,
        facet_col=facet_col,
        facet_row=facet_row,
        height=700,
    )

@app.callback(Output("bubble-plot", "figure"), [Input(d, "value") for d in indexes])
def make_figure(x, y, color, facet_col, facet_row):
    return px.scatter(
        tips,
        x=x,
        y=y,
        color=color,
        facet_col=facet_col,
        facet_row=facet_row,
        height=700,
    )


app.run_server(debug=False)
```
